#### 24.1 Freie, ungedämpfte, harmonische Schwingungen

* Schwingkreis ohne Verluste

<img src="figures/24.1_1.png" alt="24.1_1" style="width: 250px;"/>

für $t=0$: $U_C=U_0$

für $t>0$: $U_C+U_L=0$ (Maschenregel)

wobei

$$U_L = L\frac{\mathrm{d}I}{\mathrm{d}t}$$

$$U_C=\frac{1}{C}\int I \,\mathrm{d}t$$

Beides eingesetzt in die Maschenregel

$$L\frac{\mathrm{d}I}{\mathrm{d}t}+\frac{1}{C}\int I\,\mathrm{d}t=0$$

einmal differenzieren nach der Zeit

$$L\frac{\mathrm{d}^2 I}{\mathrm{d}t^2}+\frac{1}{C}I=0$$

und mit $L$ teilen ergibt

$$\ddot{I}+\frac{1}{L C}\,I = 0$$

* Physikalisches Pendel

<img src="figures/24.1_2.png" alt="24.1_2" style="width: 200px;"/>

$$\vec{M}=J_A\,\vec{\alpha}$$

Hier ist $M$ das Drehmoment, und $J_A$ das Massenträgheitsmoment bezgl. dem Aufhängepunkt $A$ und $\left|\vec{\alpha}\right|$ die Winkelbeschleunigung $\mathrm{d}^2\varphi/\mathrm{d}t^2$. Die Richtung von $\vec{\alpha}$ gibt die Drehachse an. 

$$\vec{M}=\vec{r}\times\vec{F}=\vec{s}\times\vec{F}_g$$

$$\left| \vec{M}\right|=-\left|\vec{s}\right|\,\left|\vec{F}_g\right|\,\sin\varphi$$

Das Minuszeichen beschreibt das Drehmoment entgegen den Winkel $\varphi$, also für positive Winkel $\varphi$ ist der Betrag des Drehmoments negativ und umgekehrt. Wenn Sie die Gleichung vektoriell lösen benötigen Sie diese Überlegung nicht, siehe zum Beispiel [hier](https://de.wikipedia.org/wiki/Physikalisches_Pendel). 

$$-s \cdot m \cdot g \cdot \sin \varphi = J_A \, \frac{\mathrm{d}^2\varphi}{\mathrm{d}\,t^2}$$

$$\ddot{\phi} + \frac{s\cdot m\cdot g}{J_A}\,\sin\varphi= 0\tag{24.1.1}$$

* für kleine Auslenkungswinkel 
$\varphi\le 5^\circ$ ⇨ $\sin \varphi \approx \varphi$

und Gleichung (24.1.1) vereinfacht sich zu

$$\ddot{\phi} + \frac{s\cdot m\cdot g}{J_A}\,\varphi= 0$$

* Federschwinger

<img src="figures/24.1_3.png" alt="24.1_3" style="width: 400px;"/>

Federkraftgesetz:

$$\vec{F}_F=-k \cdot x\cdot \vec{e}_x$$

$$ \vec{F} = m\,\frac{\mathrm{d}^2 x}{\mathrm{d}\,t^2}\cdot \vec{e}_x=m\, \vec{a}$$

$$m\, \ddot{x} = - k \, x $$

oder

$$\ddot{x}+\frac{k}{m}\, x = 0\tag{24.1.2}$$

Das ist eine homogene DGL 2. Ordnung mit konstanten Koeffizienten.



* Lösung der DGL: Federschwinger
  
  Welche Funktion $x=x(t)$ erfüllt die Differentialgleichung (24.1.2)?
  
  Ansatz: $x(t)=A\, e^{\,\lambda\,t}$
  
  Jetzt müssen wir die Parameter $A$ und $\lambda$ bestimmen:
  
\begin{eqnarray}
x(t)&=&A\cdot e ^{\lambda t}\\
\dot{x}(t)&=&A\cdot \lambda\cdot e ^{\lambda t}\\
\ddot{x}(t)&=&A\cdot \lambda^2\cdot e ^{\lambda t}
\end{eqnarray}

  ⇨ Einsetzen in die DGL (24.1.2)
  

\begin{eqnarray}
A\cdot \lambda^2\cdot e ^{\lambda t} + \frac{k}{m} A\cdot e ^{\lambda t} &=&0\\
A\cdot e ^{\lambda t}\cdot\left[\lambda^2 +\frac{k}{m}\right] &=&0
\end{eqnarray}
nur erfüllbar wenn $\left[\lambda^2 +\frac{k}{m}\right]=0$. Dies wird die charakteristische Gleichung genannt und wir bestimmen deren Lösungen.

$$\lambda^2=-\frac{k}{m}$$

Wir bezeichnen nun $\frac{k}{m}=\omega_0^2$, der Grund wird bald ersichtlich.

$$\lambda_{1,2}=\pm \sqrt{-\omega_0^2}=\pm\sqrt{-1}\,\omega_0=\pm i \omega_0$$

$$
\left.
\begin{array}{11}
\lambda_1=+ i \omega_0\\
\lambda_2=- i \omega_0
\end{array}
\right\} \mathrm{zwei\,Lösungen}
$$

Und einsetzten in den Ansatz

$$x(t)=A_1 e^{i\omega_0 t} +A_2 e^{-i\omega_0 t}\tag{24.1.3}$$

Bestimmung von $A_1$ und $A_2$ aus den Randbedingungen (RB)

Für die Federschwingung, $x(t=0)=x_0$ und $\dot{x}(t=0)=0$. Die letzte RB besagt, dass die Anfangsgeschwindigkeit null ist.


RB 1: $x(t=0)=x_0=A_1 + A_2$

RB 2: $\dot{x}=A_1\cdot i \omega_0\, e^{i\omega_0 t} - A_2\cdot i \omega_0\, e^{-i\omega_0 t}$
  
  für $t=0$: $\dot{x}=0$  ⇨ $A_1 i \omega_0 - A_2 i \omega_0=0$
   
   ⇨ $A_1=A_2$
   
   ⇨ Mit RB 1: $A_1=\frac{x_0}{2}$ und $A_2=\frac{x_0}{2}$

Einsetzen in die allgemeine Lösung Gl. (24.1.3):
   
$$x(t)=\frac{x_0}{2}\underbrace{\left( e^{i\omega_0 t}+e^{-i\omega_0 t}\right)}_{2\cos(\omega_0 t)}$$

$$x(t)=\frac{x_0}{2}\cdot 2\cdot \cos( \omega_0 t)$$

<div class="alert alert-success">
$$x(t)=x_0\cdot \cos( \omega_0 t)$$
</div>

für $x(t=0)=x_0$ und $\dot{x}(t=0)=0$.

* Jede andere RB führt zu anderen $A_1$ und $A_2$ ⇨ Phasenverschiebung.

Generell:

<div class="alert alert-success">
$$x(t)=x_0 \, \cos\left(\omega_0 \cdot t - \varphi_0\right)$$
</div>

  $\varphi_0$ ist die Phasenverschiebung.
  
  

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi,500)


def update(x0=1., phi0 = 0.,omega0 = 1.):
    T0=2.*np.pi/omega0
    fig = plt.figure(figsize=(7,4))
    plt.plot(t/T0,x0* np.cos( omega0*t - phi0))
    plt.text(0,.75,r'$x(t)=\cos(\omega_0 \cdot t - \varphi_0)$',fontsize=15)
    plt.ylim(-1.,1.)
    plt.xlabel(r'$t/T_0$',fontsize=15)
    plt.ylabel('Amplitude',fontsize=15)
    plt.show()

interact(update,x0=(0.,1.),phi0=(-3.,3.),omega0=(.5,3.));
```

* Interpretation
<div class="alert alert-success">
$$x(t_1)=x(t_1+T_0)$$
</div>

  $\omega_0=\frac{2\pi}{T_0}$
  
  $\omega_0=2\pi\cdot f_0$

  $f_0 = \frac{1}{T_0}$

  $\omega_0$: Kreisfrequenz
  
  $f_0$: Frequenz

  $T_0$: Periodendauer


<div class="alert alert-info">**Generelle DGL der freien ungedämpften Schwingung**
$$ \ddot{x}+\frac{k}{m}\,x = 0$$

kann geschrieben werden als

$$ \ddot{x}+\omega_0^2\,x = 0$$

mit 
$$\omega_0^2=\frac{k}{m}$$.
</div>

LC-Schwingkreis:

$$\ddot{I}+\frac{1}{LC}\,I = 0\quad\quad \omega_0^2=\frac{1}{LC}$$

Physikalisches Pendel:

$$\ddot{\varphi}+\frac{s\cdot m\cdot g}{J_A}\,\varphi = 0\quad\quad \omega_0^2=\frac{s\cdot m\cdot g}{J_A}$$


* Ein mathematisches Pendel ist ein Pendel mit einer Punktmasse an einem masselosen Faden

  <img src="figures/24.1_4.png" alt="24.1_4" style="width: 150px;"/>
 
  Das Trägheitsmoment ist
  
  $$J_A=m\cdot l^2$$ ⇨ $$\omega_0^2=\frac{s\cdot m\cdot g}{m \cdot l^2}$$ ⇨ $$\omega_0^2=\frac{g}{l}$$
  
<div class="alert alert-info">**Reduzierte Pendellänge:** äquivalent der Länge $l$ in der Schwingungsgleichung des mathematischen Pendels gleicher Schwingungsdauer des physikalischen Pendels.
$$ l_{red}=\frac{J_A}{l\cdot s}$$</div>

* Energiebilanz am Beispiel des Federschwingers

  \begin{eqnarray}
  W_{ges}&=&W_{kin}+W_{pot}\\
  &=& \underbrace{\frac{m}{2}\dot{x}^2}_{\text{kinetische und}} + \underbrace{\frac{k}{2}x^2}_{\text{potentielle Energie}}
  \end{eqnarray}
  
  $$ \dot{x}=-x_0\,\omega_0\,\sin(\omega_0\,t)=-x_0\cdot\sqrt{\frac{k}{m}}\,\sin(\omega_0\,t)$$
  
  \begin{eqnarray}
  W_{ges}&=&\frac{k}{2}\,x_0^2\,\sin^2(\omega_0\,t)+\frac{k}{2}\,x_0^2\,\cos^2(\omega_0\,t)\\
  &=& \frac{k}{2}\,x_0^2\underbrace{\left[\sin^2(\omega_0\,t)+\cos^2(\omega_0\,t)\right]}_1
  \end{eqnarray}
  
  <div class="alert alert-info">**Die Gesamtenergie ist zeitlich konstant!**
  $$W_{ges}=\frac{k}{2}\,x_0^2=\text{konst.}$$
  
  </div>
