### 26.4 Interferenz

#### 26.4.1 Interferenzbedingung

* Überlagerung zweier oder mehrerer Wellen gleicher Schwingungsfrequenz beim Zusammentreffen in einem Punkt $\vec{r}$

* geordenetes, unveränderliches Muster aus der Überlagerung
  ⇨ Bereiche minimaler und maximaler Auslenkung
  
* in der Optik

  $$ \vec{E_1}=\vec{\hat{E}}_1 \, \cos (\omega t - \vec{k}_1\cdot\vec{r} + \varphi_{01})$$
  $$ \vec{E_2}=\vec{\hat{E}}_2 \, \cos (\omega t - \vec{k}_2\cdot\vec{r} + \varphi_{02})$$


* zu einem bestimmten Zeitpunkt $t$ haben die Wellen eine *Phasendifferenz*

  $$\delta=\vec{k}_2\cdot \vec{r} - \vec{k}_1\cdot\vec{r} +\varphi_{01}-\varphi_{02}$$


* für die Interferenz darf $\delta$ nur vom Ort aber nicht von der Zeit abhängen
  
  \begin{eqnarray}
  \varphi_{01}&\ne&\varphi_{01}(t)\\
  \varphi_{02}&\ne&\varphi_{02}(t)
  \end{eqnarray}
  
  ⇨ Kohärenz
  
* Definition optischer Weg:  
  $$L=\int_P^{P'}n(s)\,\mathrm{d}s$$
  
  $n$: Brechzahl, $P$ und $P'$ Anfangs und Endpunkt des optischen Weges
  
  optischer Weg hängt neben dem Abstand $\overline{P\,P'}$ von der Brechzahl ab
  
  
* Zwei ungleiche optischen Wege, die das Licht durchlaufen kann

  $$L_1=n_1\,s_1$$
  $$L_2=n_2\,s_2$$

* Optischer Wegunterschied ⇨ *Gangunterschied*

  $$\Delta = L_2 - L_1 + \Delta'$$
  
  $\Delta'$ möglicher Phasensprung $\lambda/2$ bei Reflexion.


* Konstruktive Interferenz

  $$\delta = m\, 2\pi\quad m=0,\,\pm 1,\,\pm 2,\,\dots$$
  
  $$\Delta = m\, \lambda_0\quad m=0,\,\pm 1,\,\pm 2,\,\dots$$
  
* Destruktive Interferenz

  $$\delta = (2m +1) \, \pi\quad m=0,\,\pm 1,\,\pm 2,\,\dots$$
  
  $$\Delta = (2m+1)\, \frac{\lambda_0}{2}\quad m=0,\,\pm 1,\,\pm 2,\,\dots$$

* Zusammenhang zwischen $\delta$ und $\Delta$

  $$\delta=k_0 \,\Delta = \frac{2\pi}{\lambda_0}\Delta$$
  
  
* Überlagerung von zwei Wellen und resultierende Intensität


  Die Intensität einer monochromatischen EM-Welle ist propotional zum zeitlichen Mittel der quadrierten elektrischen Feldstärke:
  
  $$I\propto\frac{1}{T}\int_0^T\ \vec{E}^2\,\mathrm{d}t=\left<\vec{E}^2\right>_T$$

  Wir berechnen nun die Intensität zweier überlagerten monochromatischen Wellen mit einer Phasendifferenz $\delta$:
  
  \begin{eqnarray}
  E_1&=&E_0\sin(\omega t)\\
  E_2&=&E_0\sin(\omega t+\delta)\\
  I&\propto&\left<(E_1+E_2)^2\right>_T=E_0^2\left<\left(\sin(\omega t)+\sin(\omega t + \delta)\right)^2\right>_T\\
  I&\propto&E_0^2\left<\left(2\sin\left(\omega t +\frac{\delta}{2}\right)\cdot\cos\left(\frac{\delta}{2}\right)\right)^2 \right>_T\\
  I&\propto&4\,\underbrace{E_0^2 \left<\sin^2\left(\omega t +\frac{\delta}{2}\right) \right>_T}_{I}\cos^2\left(\frac{\delta}{2}\right)\\
  I&\propto&4\,I \cos^2\left(\frac{\delta}{2}\right)=4\,I \cos^2\left(\frac{\pi}{\lambda}\Delta\right)
  \end{eqnarray}
  
  
Die næchste Abbildung zeigt die Intensität für einen Gangunterschied $\delta$ und Wegunterschied $\Delta$.  

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

delta = np.linspace(0, 3.5 * np.pi,500)

I=4.*(np.cos(delta/2.))**2.

fig = plt.figure(figsize=(9,4))

plt.plot(delta,I)
plt.xlabel(r'$\delta$',fontsize=15)
plt.ylabel(u'Intensität',fontsize=15)
plt.xticks((np.arange(3)+1)*np.pi, (r'$\pi$', r'$2\,\pi$', r'$3\,\pi$'),fontsize=15)
plt.yticks((np.arange(2))*4, (r'$I_{min}$', r'$I_{max}$'),fontsize=15)
ax1 = plt.gca()
ax2 = plt.gca().twiny()
ax2.set_xlim(ax1.get_xlim())
ax2.set_xticks((np.arange(3)+1)*np.pi)
ax2.set_xticklabels((r'$\frac{\lambda}{2}$', r'$\lambda$', r'$\frac{3\lambda}{2}$'),fontsize=15)
ax2.set_xlabel(r'$\Delta$',fontsize=15)
plt.show()
```

#### 26.4.2 Interferenz an dünnen Schichten

* Überlagerungvon zwei Wellen mit zeitlich stabiler Phasendifferenz
* Lichtwelle aufspalten in zwei Anteile mit anschließender Überlagerung
* Pohlscher Interferenzversuch

<img src="figures/26.4_1.png" alt="26.4_1" style="width: 400px;"/>

* Laufzeitdifferenz zwischen $(1)$ und $(2)$
  
  ⇨ Gangunterschied
  
  ⇨ Phasenverschiebung
<p>
* optische Wege

  $L_1=2\,\cdot n_2\cdot \overline{AB}$
  
  $L_2=\overline{AD}=2\cdot\overline{AC}\,\sin\alpha$
<p>  
* Gangunterschied

  $\Delta = L_1 - L_2 + \frac{\lambda_0}{2}$
  
  $\frac{\lambda_0}{2}$ ist der Phasensprung von Strahl $(2)$ bei Relfexion am dichteren Medium.
   
  $\Delta = 2\cdot d\cdot\sqrt{n_2^2-\sin^2 \alpha} + \frac{\lambda_0}{2}$
  
  
* Interferenzmaxima: $\Delta = m \cdot \lambda_0$

  $$m\cdot\lambda_0=2\cdot d\,\cdot \sqrt{n_2^2-\sin^2 \alpha} + \frac{\lambda_0}{2}\quad m=0,1,2,\dots$$
  
* Interferenzminima: $\Delta = m \cdot \lambda_0+\frac{\lambda}{2}$

  $$m\cdot\lambda_0=2\cdot d\,\cdot \sqrt{n_2^2-\sin^2 \alpha}$$
