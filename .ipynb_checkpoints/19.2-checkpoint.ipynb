### 19.2 Kräfte des Magnetfeldes auf elektrische Ladungen


* Vergleich zum elektrostatischen Feld

  a) Bewegung von Elektronen parallel zu $\vec{E}$-Feldlinien
  
<img src="figures/19.2_1.png" alt="19.2_1" style="width: 200px;"/>

für $t=0 : \left|\vec{v}_e\right|=0$
 
   ⇨ Elektron durchläuft eine Potentialdifferenz mit Spannung $U$
   ⇨ Änderung der potentiellen Energie = Änderung der kinetischen Energie
   
   $$e\,U = \frac{m_e}{2} v_e^2$$
   
   b) Bewegung von Elektronen senktrecht zu $\vec{E}$-Feldlinien
   
<img src="figures/19.2_2.png" alt="19.2_2" style="width: 250px;"/>   

  Die Bahnkurve ist eine <b>Parabel</b>.
  
  Herleitung: Eine konstante Kraft wirkt in vertikaler Richtung, daher ist $F_y=\frac{e\,E}{m_e}=m_e\,\frac{\mathrm{d}^2y}{\mathrm{d}t^2}$. Zweimal integriert ergibt sich für die y-Koordinate $y=\frac{1}{2}\frac{e\,E}{m_e} t^2$. Die x-Position des Elektron ergibt sich ohne Kraftwirkung zu $x=v_e \,t$. Nach $t$ aufgelöst und eingesetzt ergibt sich für die y-Position $y=\frac{1}{2}\frac{e\,E}{m_e\,v_e} x^2$.
  
  
  c) Bewegte Ladungen im Magnetfeld
  
  ⇨ <b>Lorenzkraft</b>
  
  $$\vec{F_L}=Q\left[\vec{v}\times\vec{B}\right]$$
  
  $\vec{v}$ ist die Geschwindigkeit.
  
  $\vec{F_L}$ steht senkrecht zu der Ebene die $\vec{v}$ und $\vec{B}$ aufspannen.
  
  * positive Ladung $Q=+e$
  
  <img src="figures/19.2_3.png" alt="19.2_3" style="width: 250px;"/>   
  
  $\vec{B}$ zeigt aus dem Blatt heraus.
  
  * negative Ladung $Q=-e$
  
  <img src="figures/19.2_4.png" alt="19.2_4" style="width: 250px;"/>   
  
  Vorzeichen von $Q$ geht in die Richtung von $F_L$ ein.
  
  
  Betrag der Lorenzkraft:
  
  $$|\vec{F}_L|=|Q|\,|\vec{v}|\,|\vec{B}|\,\sin \sphericalangle (\vec{v},\vec{B})$$
  
  Spezialfälle:
  
  \begin{eqnarray}
  \vec{v}\parallel\vec{B}&\rightarrow& |\vec{F}_L|=0\\
  \vec{v}\perp\vec{B}&\rightarrow& |\vec{F}_L| \quad \mathrm{wird\, maximal}
  \end{eqnarray}
  
  * Anwendungen: Fadenstrahlrohr zur Bestimmung der spezifischen Ladung des Elektrons:
  $$\frac{e}{m_e}$$
  
    * Beschleunigung der Elektronen auf eine kinetische Energie 
  
  $$e U = \frac{m_e}{2}\,v_e^2$$
  
    * Einschießen in ein konstantes $\vec{B}$-Feld senkrecht zu den Feldlinien $\vec{v}\perp\vec{B}$.
    
    * Lorentzkraft übt eine Radialkraft aus
    
    $$e\,v\,B = m_e \, \frac{v_e^2}{R}$$
    
    $R$ ist der Radius der Kreisbahn.
    
    oder aufgelöst nach der spezifischen Ladung des Elektrons
    
    $$\frac{e}{m_e}=\frac{2\,U}{R^2\,B^2}$$
