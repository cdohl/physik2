## 26 Wellenoptik

### 26.1 Elektromagnetische Wellen

* Bisherige Gleichungen aus der Elektrodynamik

  Gaußsche Gesetz für elektrische Felder

  $$\oint \vec{E} \cdot \mathrm{d}\vec{A} = \frac{Q_{umg}}{\epsilon_0}$$

  Gaußsche Gesetz für magnetische Felder

  $$\oint \vec{B} \cdot \mathrm{d}\vec{A} = 0$$

  Faraday'sche Induktionsgesetz

  $$U_{ind}=\oint \vec{E}\cdot\mathrm{d}\vec{s}= - \frac{\mathrm{d}\Phi_B}{\mathrm{d}t}$$

  Ampèrsche Gesetz

  $$\oint\vec{B}\cdot\mathrm{d}\vec{s} = \mu_0 \, I_{umf}$$

  Maxwell hat gesehen, dass das Amper'sche Gesetz nicht alle Fälle der Erzeugung von Magnetfeldern beschreibt. Die Flächen S1 (hellblau) und S2 (hellrot) haben den gleichen Rand $\partial S$, so dass das Ampèresche Gesetz beider Flächen das gleiche Magnetfeld ergeben sollte. Durch die Kreisfläche $S1$ fließt der Leitungsstrom $I$ und erzeugt ein Magnetfeld. Die Draht durchstösst jedoch nicht die Fläche $S2$, und somit sollte nach dem ursprünglichen Ampèrschen Gesetz kein Magnetfeld vorliegen.
  
   <img src="figures/26.1_1.png" alt="26.1_1" style="width: 250px;"/>
   [von Wikipedia](https://de.wikipedia.org/wiki/Verschiebungsstrom).
   
  Das Ampèrsche Gesetz wurde von Maxwell erweitert:
  
  $$\oint\vec{B}\cdot\mathrm{d}\vec{s} = \mu_0 \, I_{umf}+\mu_0\epsilon_0\frac{\mathrm{d}\Phi_E}{\mathrm{d}t}$$

Damit ergeben sich die vier Maxwell'schen Gleichungen:

  $$\oint \vec{E} \cdot \mathrm{d}\vec{A} = \frac{Q_{umg}}{\epsilon_0}$$
  $$\oint \vec{B} \cdot \mathrm{d}\vec{A} = 0$$
  $$\oint \vec{E}\cdot\mathrm{d}\vec{s}= - \frac{\mathrm{d}\Phi_B}{\mathrm{d}t}$$
  $$\oint\vec{B}\cdot\mathrm{d}\vec{s} = \mu_0 \, I_{umf}+\mu_0\epsilon_0\frac{\mathrm{d}\Phi_E}{\mathrm{d}t}$$

* Kopplung zeitlich veränderlicher elektrischer und magnetischer Felder

* Kopplung der Maxwellschen Gleichungen

⇨ Magnetische Felder können auch elektrische Felder verursachen

* Elektromagnetische Wellen vor allem bei hohen Frequenzen

* Elektrische Feldstärke

  $$\vec{E}=\vec{E}_0\,\cos\left(\omega t - \vec{k}\cdot \vec{r}\right)$$

  $$\vec{E}_0\perp\vec{k}$$

* Magnetische Flussdichte

  $$\vec{B}=\vec{B}_0\,\cos\left(\omega t - \vec{k}\cdot \vec{r}\right)$$

  $$\vec{B}_0\perp\vec{k}$$
  
* für die Ausrichtung entlang der $x$-Richtung

  \begin{eqnarray}
  \vec{k}&=&k_x\,\vec{e}_x\\
  \vec{E}_0&=&\hat{E}_y\,\vec{e}_y\\
  \vec{B}_0&=&\hat{B}_z\,\vec{e}_z\\
  \end{eqnarray}
  
  Bilden ein Rechtssytem:

  $$\vec{E}_0\rightarrow\vec{B}_0\rightarrow\vec{k}\rightarrow\vec{E}_0\rightarrow\dots$$
  
  
* Wellengleichung für Ausbreitung entlang $x$-Richtung

  \begin{eqnarray}
  \frac{\partial^2 E_y}{\partial x^2}&=&\epsilon_0 \mu_0 \epsilon_r \mu_r\,
  \frac{\partial^2 E_y}{\partial t^2}\\[2ex]
  \frac{\partial^2 B_z}{\partial x^2}&=&\epsilon_0 \mu_0 \epsilon_r \mu_r\,
  \frac{\partial^2 B_z}{\partial t^2}
  \end{eqnarray}
  
  und damit ist die Phasengeschwindigkeit
  
  $$\frac{1}{c^2}=\epsilon_0 \mu_0 \epsilon_r \mu_r$$
  
* im Vakuum: $\epsilon_r=1$ und $\mu_r=1$

  ist die Vakuumlichtgeschwindigkeit $c_0$

  $$c_0=\sqrt{\frac{1}{\mu_0\epsilon_0}}\approx 299792.5\,\mathsf{km\,s^{-1}}\approx 3\cdot 10^8\,\mathsf{m\,s^{-1}}$$

* in Stoffen, hier unmagnetisch ($\mu_r=1$)
  
  $$c<c_0$$
  
  $$c=\frac{c_0}{\sqrt{\epsilon_r}}=\frac{c_0}{n}$$
  
  $n$ ist die Brechzahl
  
  $$n=\sqrt{\epsilon_r}=\frac{c_0}{c}$$
