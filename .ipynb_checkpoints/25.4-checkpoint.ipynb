### 25.4 Wellenausbreitung in Medien 

#### 25.4.1 Die gespannte Saite (Transversalwelle)

* Zugkraft $\vec{F}$

* Massenelement: $\mathrm{d}m=\rho\,A\,\mathrm{d}x$
  
  <img src="figures/25.4_1.png" alt="25.4_1" style="width: 400px;"/>

* Rücktreibende Kraft:

  \begin{eqnarray}
  \mathrm{d}\,F_{\mathrm{Rück}}&=&\mathrm{d}m\,a_y= \mathrm{d}m\,\frac{\partial^2\,y}{\partial\,t^2}=\rho\,A\,\mathrm{d}x\,\frac{\partial^2\,y}{\partial\,t^2}\tag{25.4.1}\\
  \mathrm{d}\,F_{\mathrm{Rück}}&=& F_y(x+\mathrm{d}x)-F_y(x)\\
  &=&F\,\sin(\alpha+\mathrm{d}\alpha)-F\,\sin(\alpha)
  \end{eqnarray}
  <p>
* Für kleine Auslenkungen 
  
  $$\tan(\alpha)\approx\sin(\alpha)\approx \alpha$$
  <p>
* $\mathrm{d} F_{\mathrm{Rück}}=F\,\mathrm{d}\alpha$
  <p> 
* $\tan\alpha=\lim_{\Delta x\rightarrow 0}\frac{\Delta y}{\Delta x}=\frac{\partial y}{\partial x}$  
  <p>
* $\alpha=\alpha(x)$, deshalb ist

  $$\mathrm{d}\alpha = \frac{\partial \alpha}{\partial x}\,\mathrm{d}x=\frac{\partial^2\,y}{\partial x^2}\,\mathrm{d}x$$
  
* Einsetzen in Gl. (25.4.1)

  $$F\,\frac{\partial^2\,y}{\partial x^2}\,\mathrm{d}x=\rho\,A\,\mathrm{d}x\,\frac{\partial^2\,y}{\partial\,t^2}$$
  <div class="alert alert-info">
  <p>
  $$F\,\frac{\partial^2\,y}{\partial\,x^2}=A\,\rho\,\frac{\partial^2\,y}{\partial\,t^2}$$ 
  </div>
  <p>
* Allgemeine Wellengleichung
  <p><div class="alert alert-info">
  <p>
  $$\frac{\partial^2\,y}{\partial\,x^2}=\frac{1}{c^2}\frac{\partial^2\,y}{\partial\,t^2}$$
  </div>
  <p>
* Phasengeschwindigkeit
  
  $$\frac{A\,\rho}{F}=\frac{1}{c^2}$$
  
  oder
  
  $$c=\sqrt{\frac{F}{A\,\rho}}$$
  <P>
* Diskutiere die Phasengeschwindigkeit als Funktion der Seilspannung $F$, Dichte $\rho$ und Querschnittsfläche $A$.
  

#### 25.4.2 Schallausbreitung in Gasen (Longitundinalwellen)

<img src="figures/25.4_2.png" alt="25.4_2" style="width: 400px;"/>


$u$-Auslenkung eines Teilchens

$$u(z_0+\mathrm{d}z)=u(z_0)+\frac{\partial u}{\partial z}\,\mathrm{d}z$$

⇨ wenn $u(z_0+\mathrm{d}z)\ne u(z_0)$, dann entsteht eine Druckänderung durch Änderung des Volumens

Annahmen:
* kleine Druckänderungen im Vergleich zum Umgebungsdruck
* Adiabatische Kompression und Expansion (kein Verlust durch Wärmetransport)
* Erhaltungssatz für Masse und Impuls für ein Kontinuum

$$\frac{\partial^2\,u}{\partial\,z^2}=\frac{1}{c^2}\frac{\partial^2\,u}{\partial\,t^2}$$

und die Schallgeschwindigkeit ist

$$c=\sqrt{\frac{\kappa\,R\,T}{M}}$$

wobei $\kappa$ der Adiabatenexponent ist ($\kappa=(f+2)/f$, mit $f$ die Anzahl der Freiheitsgrade), $R$ die universelle Gaskonstante ($R\approx8.31\,\mathsf{J\,mol^{-1}\,K^{-1}}$), $T$ die Temperatur in Kelvin, und $M$ die molare Masse ($\mathsf{kg\,mol^{-1}}$). 

Wir können ähnlich zur Wellengleichung der Auslenkung auch eine Wellengleichung zum Druck aufschreiben:

<p><div class="alert alert-info">
<p>
$$\frac{\partial^2\,p}{\partial\,z^2}=\frac{1}{c^2}\frac{\partial^2\,p}{\partial\,t^2}$$
<p>
</div>
