## 22 Der Wechselstromkreis
### 22.1 Erzeugung und Darstellung von Wechselströmen und -spannungen

* Wechselströme sind instationäre Ströme
* Große Bedeutung *harmonische* Wechselströme
* Kann erzeugt werden durch eine im homogenen Magnetfeld rotierende Leiterschleife:


 <img src="figures/22.1_1.png" width=400px>

 <b>Induktionsgesetz</b>
 
 $$U_{\mathrm{ind}}=-N\frac{\mathrm{d}\Phi}{\mathrm{d}t}$$
 
 $$\Phi=\vec{B}\,\cdot\,\vec{A}=B\,A\,\cos\varphi(t)$$

 gleichförmige Rotation $\varphi(t)=\omega t$
 
 \begin{eqnarray}
 U_{ind}&=&-\frac{\mathrm{d}\Phi}{\mathrm{d}t}=-B\,A\left[-\sin \omega t\right]\,\omega\\
 &=&\omega B A \sin \omega t
 \end{eqnarray}
 
 
* Momentanwert der Spannung
  $$ u(t)=U_0\sin\omega t$$
* Momentanwert des Stromes im angeschlossenen Stromkreis
  $$ i(t)=I_0\sin\left(\omega t - \varphi\right)$$
* $U_0$ und $I_0$ sind die Scheitelwerte (Spitzenwerte, Amplituden) der Spannung und des Stromes. 


* Die Kreifrequenz $\omega=2\,\pi\,f=2\pi/T$ ist die Winkelgeschwindigkeit der Leiterschleife.
* $\varphi$ ist die Phasenverschiebung der Stromstärke
* $\varphi>0$ eilt die Spannung dem Strom voraus, für $\varphi<0$ eilt der Strom der Spannung vor.



```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi,500)


def update(U0=1., I0=.6, phi = 90.):
    omega0=1.
    T0=2.*np.pi/omega0
    fig = plt.figure(figsize=(7,4))
    plt.plot(t/T0,U0* np.sin( omega0*t),'-')
    plt.plot(t/T0,I0* np.sin( omega0*t - phi/180.*np.pi),'g-.')
    plt.text(1.25,.75,r'$u(t)=U_0 \sin\omega_0 \cdot t$',fontsize=15,color='blue')
    plt.text(1.25,.55,r'$i(t)=I_0 \sin(\omega_0 \cdot t - \varphi)$',fontsize=15,color='green')
    plt.ylim(-1.,1.)
    plt.xlabel(r'$t/T_0$',fontsize=15)
    plt.ylabel('Strom/Spannung',fontsize=15)
    plt.show()

interact(update,U0=(0.,1.,0.1),I0=(0.,1.,0.1),phi=(-180.,180,15.));
```

### 22.2 Leistung elektrischer Wechselströme

* Messgeräte zeigen einen zeitlichen Mittwert, den sog. Effektivwert. 
* Kriterium: gleiche Wirkarbeit (Joule'sche Wärme) während der Zeit $T$ für Wechsel- und Gleichströme

* für Wechselstrom

  $$W=\int_0^T \underbrace{u(t) \, i(t)}_{\mathrm{ momentane \, Leistung}}\mathrm{d}t=U_0\,I_0\int_0^T\sin\omega t \, \sin(\omega t-\varphi)\mathrm{d}t\tag{1}$$

  Da $\left[\sin x\,\sin y=\frac{1}{2}(\cos(x-y)-cos(x+y))\right]$ können wir Gl.(1) so schreiben:
  
  $$=\frac{U_0\,I_0}{2}\left[\int_0^T\cos\varphi\,\mathrm{d} t-\int_0^T\cos(2\omega t-\varphi)
  \,\mathrm{d} t\right]=\frac{U_0\,I_0}{2} T\,\cos\varphi$$
  
  
  
* für Gleichstrom

  $$W=U_{eff}\,I_{eff}\,T$$
  
  bei rein ohmschen Widerständen ($R$) ist $\varphi = 0$, $U_0=R\,I_0$, und damit $U_{eff}=R\,I_{eff}$.

  ⟹ 
  
  $$\frac{U_0\,I_0}{2} T\,\cos\varphi=U_{eff}\,I_{eff} \,T$$
  
  $$I_{eff}=\frac{I_0}{\sqrt{2}}\approx 0.707 \, I_0$$
  $$U_{eff}=\frac{U_0}{\sqrt{2}}\approx 0.707 \, U_0$$
  
  ansonsten: Leistungsfaktor $\cos\varphi$.
  

## 22.3 Wechselstromwiderstände

Wechselstromkreis: $U$, $R$, $C$, $L$

$R$: Wirkwiderstand

$X_L$: Induktiver Blindwiderstand

$X_C$: kapazitiver Blindwiderstand


* Folge von $L$ und $C$ ⟹ Phasenverschiebung $\varphi$ zwischen $u(t)$ und $i(t)$

* deswegen Rechnung in der komplexen Zahleneben (Zeigerdiagramm).

  \begin{eqnarray}
  u(t)&=&U_0\sin\omega t\\
  i(t)&=&I_0\sin(\omega t -\varphi)
  \end{eqnarray}

  ⟹ Übergang ins Komplexe $j=\sqrt{-1}$, $z=r\,e^{j\varphi}=x+j y$ und $e^{j\varphi}=\cos\varphi+j\sin\varphi$.
 
  \begin{eqnarray}
  u(t)&=&U_0e^{j\omega t}\\
  i(t)&=&I_0e^{j\omega t -\varphi}
  \end{eqnarray}

 
 * Komlexer Widerstand (Impedanz): $Z_l=j\omega L$ und $Z_C=\frac{1}{j\omega C}=\frac{j}{j^2\omega C}=-\frac{j}{\omega C}$
 
 * Blindwiderstände (komplexer Anteil der Impedanz): $X_L=\omega L$ und $X_C=\frac{1}{\omega C}$.
 
 * Wirkwiderstand (reller Anteil der Impedanz): $R$
 
 * Reihenschaltung von $R$, $L$ und $C$:
 
 <img src="figures/22.3_1.png" width=400px>
 
  Die komplexen Widerstände werden addiert:
 
  $$Z=R+j\,\omega L - \frac{j}{\omega C}=R+j\left(\omega L - \frac{1}{\omega C}\right)$$
 
  Die Zeiger in der Abbildung unten beschreiben die Addition der 3 Widerstände in der komplexen Ebene:
 
  <img src="figures/22.3_2.png" width=400px>
  
  Der Scheinwiderstand ist der Betrag von $Z$ (Länge des Zeigers) mit $U=\left|Z\right|I$ :
  
  $$\left|Z\right|=\sqrt{R^2+\left( \omega L - \frac{1}{\omega C}\right)^2}$$
  
  Phasenverschiebung $\varphi$ der Gesamtspannung gegenüber dem Strom:
 
  $$\tan \varphi=\frac{X_L-X_C}{R}=\frac{\omega L-\frac{1}{\omega C}}{R}$$
  
  Die Effektivwerte der Spannung am Widerstand $U_R=R I$, Spule $U_L=X_L I$ und Kondensator $U_C=X_C I$ addieren sich vektoriell:
  
  $$U=\sqrt{{U_R}^2+\left(U_L-U_C\right)^2}$$
