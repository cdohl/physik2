#### 16.2 Kräfte zwischen Ladungen (Coloumbgesetz)

* Ladungen mit unterschiedlichen Vorzeichen ziehen sich an, mit gleichen Vorzeichen &#8594; Abstoßung

* Zwischen zwei Ladungen $Q_1$ und $Q_2$ nimmt die Kraftwirkung mit dem Abstand ab.

* Beschreibung durch das Coloumbgesetz

Kraft die Ladung $Q_1$ auf Ladung $Q_2$ ausgeübt, wenn sich Ladung $Q_1$ im Ursprung befindet.



<img src="figures/16.2_1.png" alt="16.2_1" style="width: 200px;"/>


$$\vec{F}_C=\frac{1}{4\pi\epsilon_0} \cdot \frac{Q_1 \cdot Q_2}{r^2}\cdot \vec{e}_r$$

$\vec{e}_r$ verbindet $Q_1$ mit $Q_2$. Der Vektor auf $Q_2$ ist $\vec{r}=r\cdot \vec{e}_r$.


```python
%matplotlib notebook
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

r = np.linspace(0.01, 10, 1000)
fig, ax = plt.subplots(1, figsize=(6, 3))
fig.suptitle("Coloumbgesetz in einer Dimension")

def update_plot(dist, q1, q2):
    ax.clear()
    ### Colombgesetz ohne 1/(4 pi eps)
    fc = q1*q2/r**2.
    fc_dist = q1*q2/dist**2
    
    mycolor = {True: "r", False: "b"}
    mypm = {True: "w+", False: "w_"}
    fch, = ax.plot(r, fc)
    ax.axhline(0, color='black', lw=1.)
    ax.axvline(0, color='black', lw=1.)
    
    ax.arrow(dist, fc_dist, fc_dist, 0, lw=2., head_width=1., head_length=.5, fc=mycolor[q2>0], ec=mycolor[q2>0])
    q2h, =ax.plot(dist,fc_dist,'o', markersize=np.abs(q2)*5., color=mycolor[q2>0])
    q2hw, =ax.plot(dist,fc_dist,mypm[q2>0], markersize=np.abs(q2)*5.)
    q1h, =ax.plot(0,0,'o', markersize=np.abs(q1)*5., color=mycolor[q1>0])
    q1hw, =ax.plot(0,0, mypm[q1>0], markersize=np.abs(q1)*5.)
     
    ax.text(-1.,2.,'$Q_1$',fontsize=15)    
    ax.text(dist,fc_dist+np.sign(fc_dist)*3,'$Q_2$',fontsize=15)

    ax.set_xlim(r[0], r[-1])
    ax.legend([fch, (q1h,q1hw), (q2h,q2hw)], ["$F_C$={:.1f}".format(fc_dist), "$Q_1$={}".format(q1),
                                              "$Q_2$={}".format(q2)], loc=1)
    ax.set_xlabel('Abstand (a.u.)')
    ax.set_ylabel('Kraft (a.u.)')
    
    ax.set_xlim(-2., 10.)
    xticks = ax.xaxis.get_major_ticks()
    xticks[0].label1.set_visible(False)
    ax.set_ylim(-10., 10.)
    plt.tight_layout()
    plt.show()


dist = widgets.FloatSlider(min=.1, max=8, value=1, step=0.1, description='Abstand')
q1 = widgets.FloatSlider(min=-5, max=5, value=1, step = 0.5, description='$Q_1$')
q2 = widgets.FloatSlider(min=-5, max=5, value=-1, step=0.5, description='$Q_2$')

widgets.interactive(update_plot, dist=dist, q1=q1, q2=q2)
```

oder für den Abstandsvektor $\vec{r}$ (Vakuum):

$$\vec{F}_C=\frac{1}{4\pi\epsilon_0} \cdot \frac{Q_1 \cdot Q_2}{r^3}\cdot \vec{r}$$

*  Elektrische Feldkonstante oder Influenzkonstante $\epsilon_0=8.854 \cdot 10^{-12}\,\mathsf{A\,s}$V$^{-1}\,$m$^{-1}$

* Vergleiche Coloumbgesetz mit Gravitationsgesetz

<center>
$F_C=K_1\, Q_1 Q_2 \frac{1}{r^2}$ vs. $F_G=K_2\, m_1 m_2 \frac{1}{r^2}$
</center>
<p>
<center>
$$K_1 = 9.0\cdot 10^9\,\mathsf{N\,C^2\,kg^{-2}} \quad K_2=6.7\, \mathsf{N\,m^2\,kg^{-2}}$$
</center>

$$\frac{F_C}{F_G} \approx 10^{36} !!!$$
