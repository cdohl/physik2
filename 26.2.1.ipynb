#### 26.2 Das Huygensche Prinzip

#### 26.2.1 Formulierung und Demonstration

*Jeder Punkt einer bestehenden Wellenfront (Fläche konstanter Phase) ist Ausgangspunkt einer kugelförmigen Elementarwelle, die die gleiche Ausbreitungsgeschwindigkeit und Frequenz wie die ursprüngliche Wellenfront hat. Die Einhüllende aller überlagerten Elementarwellen ergibt die Wellenfront zu einem späteren Zeitpunkt.*

Christian Huygens (1629-1695)



#### 26.2.2 Das Reflektionsgesetz


* Reflexion erfolgt immer dann, wenn sich an einer Grenzfläche die Phasengeschwindigkeit ändert.

* Reflexion zwischen einem optisch "dünnen" Medium $(c_1,n_1)$ zu einem optisch dichteren Medium $(c_2,n_2)$.
 
  $$c_1>c_2$$
  
  $$n_1<n_2$$
  
  $$n_1=\frac{c_0}{c_1}$$
  
  $$n_2=\frac{c_0}{c_2}$$
  
* Betrachtung einer ebenen Welle, bei der Reflexion werden nach und nach Elementarwellen ausgesendet.

* Herleitung der Reflexionsgesetzes
 zwei Dreiecke: $A,B,C$ und $A,B,D$
 
 $$\sin \alpha = \frac{\overline{BC}}{\overline{AB}}=\frac{c_1\cdot\Delta t}{\overline{AB}}$$
 
 $$\sin \alpha' = \frac{\overline{AD}}{\overline{AB}}=\frac{c_1\cdot\Delta t}{\overline{AB}}$$
 
 $$\frac{\sin \alpha}{\sin \alpha'}=1$$
 
 $$\rightarrow \alpha=\alpha'$$
 
 oder *Einfallswinkel ist Ausfallswinkel*.

* Herleitung des Brechungsgesetzes

<img src="figures/26.2_4.png" alt="26.2_4" style="width: 300px;"/>

  zwei Dreiecke: $A,B,C$ und $A,B,D$
  
  $$\sin \alpha = \frac{\overline{BC}}{\overline{AB}} = \frac{c_1\cdot\Delta t}{\overline{AB}}$$
  
  $$\sin \beta = \frac{\overline{AD}}{\overline{AB}}=\frac{c_2\cdot\Delta t}{\overline{AB}}$$
 
  $$\frac{\sin \alpha}{\sin \beta} = \frac{\overline{BC}}{\overline{AD}}$$
  
* Snellius-Brechungsgesetz
  $$\frac{\sin \alpha}{\sin \beta} = \frac{c_1}{c_2} = \frac{n_2}{n_1}$$
  
  Übergang in den Bereich geringerer Phasengeschwindigkeit (höherer Brechzahl)
  
  ⇨ Welle wird zum Lot hin gebrochen
  
  
#### 26.2.4 Totalreflexion

* Übergang vom optisch dichteren in optisch dünnere Medium $n_2 > n_1$

<img src="figures/26.2_5.png" alt="26.2_5" style="width: 300px;"/>


  $$\frac{\sin \alpha}{\sin \beta} = \frac{c_1}{c_2} = \frac{n_2}{n_1}$$
  
  Grenzwinkel der Totalreflexion: $\beta_T$
  
  
  $$\sin\alpha=1$$
  
  ⇨ $\alpha = 90^\circ$
  
  $$\sin \beta_T = \frac{n_1}{n_2}$$
  
* für $\beta>\beta_T$ erfolgt vollständige Reflexion

* Einige Stoffe: an der Grenzfläche zu Luft ($n_1=1$)

| Stoff | $n$ | $\beta_T$|
|-------|-----|----------|
| Wasser|1.33|$48^\circ$|
| Glas|1.5|$42^\circ$|
| Diamant|2.4|$25^\circ$|
