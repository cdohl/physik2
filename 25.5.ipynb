### 25.5 Stehende Wellen

#### 25.5.1 Schwingende Saiten

* Phasengeschwindigkeit sei konst.

  $$c=\sqrt{\frac{F}{A\,\rho}}=\mathrm{konst.}$$
  
* Überlagerung von zwei Wellen

  * gleiche Phasengeschwindigkeit
  * Entgegengesetzte Ausbreitungsrichtung (Realisierung z.B. durch Reflexion)
  * Amplitude sei gleich
  
* bei Reflexion am festen Ende
  ⇨ Phasensprung  von $\pi$ bzw. $\lambda/2$
  
* bei Reflexion am losen Ende
  ⇨ kein Phasensprung
  
\begin{eqnarray}
A_1 &=& \hat{A} \cos \left( \omega\,t -k\,z +\varphi_0\right)\quad&\mbox{pos. z-Richtung}&\\
A_2 &=& \hat{A} \cos \left( \omega\,t +k\,z \right)\quad&\mbox{neg. z-Richtung}&\\[10pt]
A_{ges}&=&A_1+A_2\\
&=& \underbrace{2\,\hat{A}\,\cos\left(k\,z-\frac{\varphi_0}{2}\right)}_{\mbox{räumlich variierende Amplitude}}\cdot\underbrace{\cos\left(\omega\,t + \frac{\varphi_0}{2}\right)}_{\mbox{zeitliche Periodizität}}
\end{eqnarray}


* Momentanbilder einer stehenden Welle $L=\lambda/2$
  * beidseitig eingespannt 
  * an den Rändern ist ein Knoten
  * in der Mitte ein Bauch
  * für die Zeiten $t_1$, $t_2$ und $t_3$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

L = 1.
T = 1.
omega = 2*np.pi/T
z = np.linspace(0., L, 100)
phi=-np.pi/2.
n = 1
l = 2.*L/n #lambda
k=2.*np.pi/l
fig = plt.figure(figsize=(7,4))
t=0.;w=np.cos(omega*t)*np.cos(k*z+phi);plt.plot(z/L,w,'r')
t=.25;w=np.cos(omega*t)*np.cos(k*z+phi);plt.plot(z/L,w,'g')
t=.5;w=np.cos(omega*t)*np.cos(k*z+phi);plt.plot(z/L,w,'b')
xposition = [0., 1.]
for xc in xposition:
    plt.axvline(x=xc, color='k', linestyle='--')
plt.ylim(-1.1,1.1)
plt.xlabel(r'$z$',fontsize=15)
plt.ylabel('Amplitude',fontsize=15)
plt.gca().axes.get_xaxis().set_ticks([])
plt.gca().axes.get_yaxis().set_ticks([])
plt.gca().annotate("", xy=(0,-0.7), xytext=(1.,-0.7),
            arrowprops={'arrowstyle': '<->'}, va='center')
plt.gca().annotate("Bauch", xy=(0.5,1.), xytext=(.5,.2),
            arrowprops={'arrowstyle': '->'}, va='center',fontsize=15)
plt.gca().annotate("Knoten", xy=(1.,0.), xytext=(.8,.8),
            arrowprops={'arrowstyle': '->'}, va='center',fontsize=15)
plt.text(.43,-0.6,r'$L=\lambda/2$',fontsize=15)
plt.text(0.1,0.5,r'$t_1$',fontsize=15,color='r')
plt.text(.1,.1,r'$t_2$',fontsize=15,color='g')
plt.text(.1,-0.6,r'$t_3$',fontsize=15,color='b')

plt.show()

```

Hier können Sie die Anregungsfrequenz durch die Mode $n$ vorgeben und die Zeit $t$ variieren. 

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

L = 1.
T = 1.
omega = 2*np.pi/T
z = np.linspace(0, L, 100)
phi=-np.pi/2.

def update(t=0., n=1):
    l = 2*L/n #lambda
    k=2*np.pi/l
    w=np.cos(omega*t)*np.cos(k*z+phi)
    fig = plt.figure(figsize=(7,4))
    plt.plot(z/L,w,'k')
    plt.ylim(-1.1,1.1)
    plt.xlabel(r'$z/L$',fontsize=15)
    plt.ylabel('Amplitude',fontsize=15)
    plt.axhline(y=0, color='k')
    xposition = [0., 1.]
    for xc in xposition:
        plt.axvline(x=xc, color='k', linestyle='-')
    plt.show()

interact(update,t=widgets.FloatSlider(value=0.,min=0.,max=1,step=0.05,description="t/T"), n=(1,10,1));
```

* Die geometrische Größe $L$ legt die Wellenlänge fest für ein beidseitig eingespanntes Seil

  ⇨ 
  $$ L=n\,\frac{\lambda_n}{2},\quad n=1,2,3,4,\dots$$
  
  ⇨ da $c=f_n\,\lambda_n$
  $$ f_n=\frac{c}{\lambda_n}=\frac{c}{2L}\,n$$
  
  Bedingung für die Anregungsfrequenzen bei denen stehende Wellen bei festem $L$ angeregt werden können.

#### 2-dimensionale Stehwellen und Chladnische Klangfigur

Für feste Enden hat die 1-dimensionale Wellengleichung für ein Seil der Länge $L$

$$\frac{\partial^2 z}{\partial x^2} = \frac{1}{c}\frac{\partial^2 z}{\partial t^2}$$

die Lösung 

$$z_n(x,t)=A_n\sin\left(\frac{n\pi x}{L}\right)\cos\left(\omega_n t\right)\quad n=1,2,\dots\quad.$$

Die Frequenzen der Stehwelle sind

$$\omega_n=c\,k_n=\frac{c\,n\pi}{L}\quad .$$

Die Wellengleichung in 2 Dimensionen lautet

$$\frac{\partial^2 z}{\partial x^2} \, \frac{\partial^2 z}{\partial y^2}= \frac{1}{c}\frac{\partial^2 z}{\partial t^2}$$

Die Randbedingungen sind jetzt 

$x=0, x=L_x\,\rightarrow\, z=0$

$y=0, y=L_y\,\rightarrow\, z=0\quad .$

Wir benötigen nun zwei Zahlen $m,n$ um die Moden in $x$- und $y$-Richtung zu beschreiben. 

Wir können die Lösung der 2-dimensionalen Wellengleichung aufschreiben:

$$z_{m,n} = A_{m,n}\,\sin\left(\frac{m\pi x}{L_x}\right)\,\sin\left(\frac{n\pi x}{L_y}\right)\,\cos\left(\omega_n t\right)$$

und die Frequenz ist

$$\omega_{m,n}=c\,k_{m,n}=c\sqrt{\left(\frac{m\pi}{L_x}\right)^2+\left(\frac{n\pi}{L_y}\right)^2}\quad .$$
