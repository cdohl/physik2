### 16.4 Elektrostatisches Potential und Spannung

* Coulombkraft ist eine konservative Kraft

Das elektrische Feld ist wirbelfrei, d.h. 
das Integral über einen geschlossenen Weg:


$$ \oint \vec{F}\cdot \mathrm{d}\vec{r} = 0 $$

* Definition von einem elektrischen Potential an Position $\vec{r}_1$ ist möglich

$$E_{pot}=-\int_\infty^{\vec{r}_1} \vec{F} \cdot \mathrm{d}\vec{r} $$

Nullpunkt von $E_{pot}$ liegt bei $\infty$.

* Elektrisches Potential einer Punktladung $Q_2$ im Feld einer Punktladung $Q_1$

\begin{align}
E_{pot}&=&-\int_\infty^{\vec{r}_1} \vec{F_c} \cdot \mathrm{d}\vec{r}\\
&=&-\int_\infty^{\vec{r}_1} Q_2 \cdot \vec{E} \cdot \mathrm{d}\vec{r}\\
&=&-Q_2\int_\infty^{\vec{r}_1} \vec{E} \cdot \mathrm{d}\vec{r}\\
...\\
E_{pot}&=& \frac{1}{4 \pi\epsilon}\frac{Q_1 \cdot Q_2}{r_1}
\end{align}


* Verlauf der potentiellen Energie


```python
%matplotlib notebook
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

r = np.linspace(0.01, 10, 1000)
fig, ax = plt.subplots(1, figsize=(6, 3))
fig.suptitle("Potentielle Energie")

def update_plot(q1, q2):
    ax.clear()
    ### Colombgesetz ohne 1/(4 pi eps)
    Epot = q1*q2/r**2.
    mycolor = {True: "r", False: "b"}
    mypm = {True: "w+", False: "w_"}
    ax.text(5,3.,r'$E_{pot}=k \,\frac{Q_1 \cdot Q_2}{r}$',fontsize=15)
    fch, = ax.plot(r, Epot)
    ax.axhline(0, color='black', lw=1.)
    ax.axvline(0, color='black', lw=1.)
    q1h, =ax.plot(0,0,'o', markersize=np.abs(q1)*5., color=mycolor[q1>0])
    q1hw, =ax.plot(0,0, mypm[q1>0], markersize=np.abs(q1)*5.)
     
    ax.text(-1.,2.,'$Q_1$',fontsize=15)   
    ax.set_xlim(r[0], r[-1])
    
    ax.set_xlabel('Abstand (a.u.)')
    ax.set_ylabel('$E_{pot}$ (a.u.)')
    
    ax.set_xlim(-2., 10.)
    #xticks = ax.xaxis.get_major_ticks()
    #xticks[0].label1.set_visible(False)
    ax.set_ylim(-10., 10.)
    ax.tick_params(labelbottom='off',labelleft='off',which='both', bottom='off',top='off', left='off'  )   
    plt.tight_layout()
    
    plt.show()


q1 = widgets.FloatSlider(min=-5, max=5, value=1, step = 0.5, description='$Q_1$')
q2 = widgets.FloatSlider(min=-5, max=5, value=-1, step=0.5, description='$Q_2$')

widgets.interactive(update_plot, q1=q1, q2=q2)
```

* elektrisches Potential der Ladung $Q_1$

$$\varphi(\vec{r})=-\int_\infty^{\vec{r}_1} \vec{E} \cdot \mathrm{d}\vec{r}$$

$$\varphi(\vec{r})=\frac{E_{pot}}{Q_2}$$

* elektrische Spannung zwischen zwei Punkten ⇨ Potentialdifferenz

$$\varphi(\vec{r}_B)-\varphi(\vec{r}_A)=U$$

$U$ ist die Spannung.

\begin{align}
\varphi(\vec{r_B})&=&-\int_\infty^{\vec{r}_B} \vec{E} \cdot \mathrm{d}\vec{r}\\
\varphi(\vec{r_A})&=&-\int_\infty^{\vec{r}_A} \vec{E} \cdot \mathrm{d}\vec{r}
\end{align}

Eingesetzt in die Formel für die Spannung ergibt sich

$$U=-\int_{\vec{r}_A}^{\vec{r}_B} \vec{E} \cdot \mathrm{d}\vec{r} $$

Negative Ladung wird zur positiven Spannung beschleunigt!
