### 26.3 Polarisation


* EM-Wellen sind Transversalwellen

 $$\vec{E}=\vec{E_0}\,\cos (\omega t - \vec{k}\cdot \vec{r})$$
 
 $\vec{E}_0\perp \vec{k}$ analog $\vec{B}_0\perp \vec{k}$
<p> 

* unpolarisiertes Licht:
    
  wenn jede Richtung von $\vec{E_0}$ senkrecht zu $\vec{k}$ in gleichen Anteilen auftritt

<p>
* Beispiel für unpolarisierte Strahlung
  
  ⇨ Glühlampe (spontane Emission)
<p>  
* Lichtausbreitung in $x$-Richtung
  
  $$\vec{k} = k_x \vec{e}_x$$
  
  $\vec{E}_0$ liegt in der y-z-Ebene
  
* gibt es eine Vorzugsrichtung von $\vec{E_0}$ bzw. $\vec{B}_0$

  ⇨ Polarisation
  
  Bsp.: Mikrowelle, Laserstrahlung, Dipolstrahlung
  
  
* durch Polarisatoren kann man unpolarisierte Strahlung in polarisierte überführen

<p>

* für $\vec{k}=k_x\vec{e}_x$

  $$\vec{E}_0=E_y \vec{e}_y +E_z \vec{e}_z$$
  
  $$E_y=\hat{E}_y\,\cos(\omega t - k_x\,x+\varphi_{0y})$$
  
  $$E_z=\hat{E}_z\,\cos(\omega t - k_x\,x+\varphi_{0z})$$

  <img src="figures/26.3_1.png" alt="26.3_1" style="width: 250px;"/>
  
* $\omega$ ist für beide Komponenten identisch

  ⇨ Analyse der Feldstärken
  
  $$\hat{E}_y : \hat{E}_z\quad \mathrm{und}\quad \varphi_{0y}-\varphi_{0z}$$
  
  ⇨ ist analog zu Lissajous-Figuren mit Frequenzverhältnis $1:1$
<p>  
* linear polarisiertes Licht

  $$\varphi_{0y}-\varphi_{0z}=0^\circ\quad\varphi_{0y}=\varphi_{0z}$$

   <img src="figures/26.3_2.png" alt="26.3_2" style="width: 250px;"/>
   
  ⇨ eine einzige Schwingungsrichtung von $\vec{E}_0$.
  
  
* zirkular polarisiertes Licht

  $$\hat{E}_z=\hat{E}_y\quad\varphi_{0y}-\varphi_{0z}=90^\circ\quad\mathrm{bzw.}\quad 270^\circ$$

  <img src="figures/26.3_3.png" alt="26.3_3" style="width: 250px;"/>
  
  für einen festen Ort $x$ läuft der Vektor $\vec{E}_0$ mit der Zeit periodisch um.
<p>    
* jeder andere Fall liefert elliptisch polarisiertes Licht
