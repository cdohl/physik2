### 25.3 Wellenausbreitung im Raum

* Wellenfronten: Verbindung aller Punkte mit gleichem Auslenkungszustand, d.h. mit $\varphi=\mathrm{konst.}$
  * Punkterregung im Raum
  ⇨  Kugeloberfläche ⇨ Kugelwellen
  * Punkterregung in der Ebene
  ⇨  Kreiswellen

  * Wellenfronten sind Ebenen
  ⇨  ebene Wellen

  Flächen konstanter Phase einer ebenen Welle
  <img src="figures/25.3_3.png" alt="25.3_3" style="width: 200px;"/>
  <p>
* Zur Beschreibung der Ausbreitung im Raum wir der Wellenzahlvektor $\vec{k}$ benutzt. $\vec{k}$ gibt die Ausbreitungsrichtung der Welle an.
  $$\vec{k}=\underbrace{\left|\vec{k}\right|}_{\frac{2\pi}{\lambda}} \, \vec{e}_k$$
  
  $$\vec{k}=k_x\,\vec{e}_x+k_y\,\vec{e}_y+k_z\,\vec{e}_z$$
  
  $$\vec{r}=x\,\vec{e}_x+y\,\vec{e}_y+z\,\vec{e}_z$$
  
  $$u(t,r)=\hat{u}\cdot\sin\left(\omega t - \vec{k}\cdot \vec{r} + \varphi_0\right)$$
  
  <img src="figures/25.3_4.png" alt="25.3_4" style="width: 250px;"/>
  
  Bsp.: Ausbreitung entlang $x$-Richtung:
  Fläche konstanter Phase
  $$\omega t - k_x x + \varphi_0=\mathrm{konst.}$$
  <p>
  
  Allgemein:
  
  <img src="figures/25.3_5.png" alt="25.3_5" style="width: 250px;"/>
  
  \begin{eqnarray}
  A(\vec{r},t)&=&\hat{A}\,\sin\left(\omega\,t - \vec{k}\cdot\vec{r}+\varphi_0 \right)\\
  A(\vec{r},t)&=&\hat{A}\,\sin\left(\omega\,t - k_x\cdot x - k_y\cdot y- k_z\cdot z+\varphi_0 \right)\\
  \end{eqnarray}
  
  **Komplexe Notation:**
  
  $$A(\vec{r},t)=\hat{A} e^{i\left(\omega\,t-\vec{k}\cdot\vec{r}+\varphi_0\right)}$$
  
  Wenn die Amplitude eine Richtung hat:
  
  $$\vec{A}(\vec{r},t)=\vec{\hat{A}} e^{i\left(\omega\,t-\vec{k}\cdot\vec{r}+\varphi_0\right)}$$
  
  **Longitudinalwellen:**
  
  Amplitude ist parallel zur Ausbreitungsrichtung, $\vec{\hat{A}}\parallel \vec{k}$. 
  
  **Transversalwellen:**
  
  Amplitude ist senkrecht zur Ausbreitungsrichtung, $\vec{\hat{A}}\bot \vec{k}$. 
  
  
  
  
* Intensität für verschiedene Fälle in Abhängigkeit vom Abstand $|\vec{r}|$ zum Erregerzentrum:
  * ebene Welle $I=\mathrm{konst.}$
  * Kugelwelle $I\propto r^{-2}$
  * Kreiswelle $I\propto r^{-1}$
