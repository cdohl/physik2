### 25.6 Dopplereffekt

#### 25.6.1 Beobachter in Ruhe

* a) Quelle bewegt sich auf Beobachter zu

  $v_Q$: Quellengeschwindigkeit
  
  $c$: Phasengeschwindigkeit der Welle im Medium
  
  * für $v_Q<c$:
    
    Verkürzung der Wellenlänge
  
    \begin{eqnarray}
    \lambda&=&\lambda_0 - v_Q\cdot T_0\quad T_0=\frac{1}{f_0}\\
    \lambda&=&\lambda_0 - \frac{v_Q}{f_0}=\lambda_0\left(1-\frac{v_Q}{c}\right)
    \end{eqnarray}
  
    Frequenz beim Betrachter:
    
    $$f=\frac{c}{\lambda}=\frac{\lambda_0 \, f_0}{\lambda}=\frac{f_0}{1-\frac{v_Q}{c}}$$
  
  * für $v_Q>c$ (Überschallgeschwindigkeit):
    
    ⇨ Machsche Kegel
    
    <img src="figures/25.6_1.png" alt="25.6_1" style="width: 250px;"/>
    
    $$\sin \alpha = \frac{c\cdot t_{12}}{v_Q\cdot t_{12}}$$
    
    $$\sin \alpha = \frac{c}{v_Q} = \frac{1}{M}$$
    
    $M$: Machzahl
  
  
* b) Quelle bewegt sich weg vom Beobachter

  ⇨ Zunahme der Wellenlänge
  
  \begin{eqnarray}
  \lambda &=& \lambda_0 + v_Q \cdot T_0\\
  f&=&\frac{f_0}{1+\frac{v_Q}{c}}<f_0
  \end{eqnarray}
  
  Verringerung der Frequenz
  
  
#### 25.6.1 Quelle in Ruhe ($v_Q=0$)

* c) Beobachter bewegt sich auf die Quelle zu mit $v_B$
  
  für $v_B=0$ passieren in $\Delta t$ 
  
  $$ n_0 = \frac{c \cdot \Delta t}{\lambda_0}$$
  
  Wellenberge im Ohr.
  
  für $v_B>0$ ändert sich die Anzahl der Wellenberge im Zeitintervall $\Delta t$
  
  $$\Delta n = \frac{v_B \cdot \Delta t}{\lambda_0}$$
  
  ⇨
  
  $$f=f_0\,\left(1+\frac{v_B}{c}\right)>f_0$$
  
  
* d) Beobachter bewegt sich von der Quelle weg

  $$f=f_0\,\left(1-\frac{v_B}{c}\right)<f_0$$
